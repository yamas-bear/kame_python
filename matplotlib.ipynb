```python
df = pd.read_csv("dataset/tmdb_5000_movies.csv")
```

```python
df.head()
```

```python
df = df[(~df['release_date'].isna())]
```

```python
df['year'] = df['release_date'].apply(get_year)
```

```python
def get_year(release_date):
    return int(release_date[:4])
```

```python
df['year']
```

```python
plt.plot(df.groupby('year').count()['id'])
plt.xlabel('year')
plt.ylabel('count')
plt.title('Yearly change in the number of movies')
```

```python
df = df[df['revenue']!=0]
plt.plot(df.groupby('year').median()['revenue'],label='medisn')
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
plt.scatter(df['revenue'],df['budget'],alpha=0.2)
```

```python
df = df[(df['revenue']!=0) & (df['budget'] !=0)]
lang_group_df = df.groupby('original_language').mean()[['budget','revenue','vote_count']]
plt.scatter('revenue','budget',s='vote_count',data=lang_group_df,c=np.arange(len(lang_group_df)))
for idx,row in lang_group_df.iterrows():
    plt.annotate(idx,(row['revenue'],row['budget']))
```
